```python
df.dtypes
```

```python
df.head()
```

**Tratando valores faltantes**

```python
#consultando linhas com valores faltantes
df.isnull().sum()
```

```python
#substituindo os valores nulos pela média
df["Vendas"].fillna(df["Vendas"].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
#substituíndo os valores nulos por zero
df["Vendas"].fillna(0, inplace=True)
```

```python
#apagando as linhas com valores nulos
df.dropna(inplace=True)
```

```python
#apagando as linhas com valores nulos com base apenas em 1 coluna
df.dropna(subset=["Vendas"], inplace=True)
```

```python
#removendo linhas que estejam com valores faltantes em todas as colunas
df.dropna(how="all", inplace=True)
```

**Criando colunas novas**

```python
#criando a coluna de receita
df["Receita"] = df["Vendas"].mul(df["Qtde"])
```

```python
df.head()
```

```python
#achando valor da coluna quantidade caso não tivesse
df["Receita/Vendas"] = df["Receita"] / df["Vendas"]
```

```python
#qual a receita total?
df6["Valor Venda"].sum()
```

```python
#qual o custo total?
df6["custo"] = df6["Custo Unitário"].mul(df6["Quantidade"])#criando a coluna custo
```

```python
df6.head(1)
```

```python
#qual o custo total?
round(df6["custo"].sum(), 2)
```

```python
#agora que temos a receita e custo e o total, podemos achar o lucro total
#vamos criar uma coluna de Lucro que será Receita - Custo
df6["lucro"] = df6["Valor Venda"] - df6["custo"]
```

```python
df6.head(1)
```

```python
#total lucro
round(df6["lucro"].sum(), 2)
```

```python
#criando uma coluna dom total de dias para enviar o produto
df6["Tempo_envio"] = df6["Data Envio"]- df6["Data Venda"]
```

```python
df6.head(1)
```

**Agora, queremos saber a média do tempo de envio para cada Marca, e para isso precisamos transformar a coluna Tempo_envio em númerica**

```python
#extraindo apenas os dias
df6["Tempo_envio"] = (df6["Data Envio"] - df6["Data Venda"]).dt.days
```

```python
df6.head(1)
```

```python
#verificando o tipo da coluna Tempo_envio
df6["Tempo_envio"].dtype
```

```python
#média do tempo de envio por Marca
df6.groupby("Marca")["Tempo_envio"].mean()
```
